```python
import numpy as np

# Given arrays
x = np.array([[1, 2], [3, 5]])
y = np.array([[5, 6], [7, 8]])
v = np.array([9, 10])
w = np.array([11, 12])

array_sum = x + y
vector_sum = v + w
print("Sum of x and y:")
print(array_sum)
print("\nSum of v and w:")
print(vector_sum)

print("\n")

array_difference = x - y
vector_difference = v - w
print("Difference of x and y:")
print(array_difference)
print("\nDifference of v and w:")
print(vector_difference)

print("\n")

x_multiply = x * 3
v_multiply = v * 2
print("Array x multiplied by 3:")
print(x_multiply)
print("\nVector v multiplied by 2:")
print(v_multiply)

print("\n")

# Square each element
x_squared = x ** 2
v_squared = np.square(v)
print("Squared elements of x:")
print(x_squared)
print("\nSquared elements of v:")
print(v_squared)

print("\n")

# Calculate the dot product
dot_product1 = np.dot(v, w)
dot_product2 = np.dot(x, v)
dot_product3 = np.dot(x, y)
print("Dot Product of v and w:")
print(dot_product1)
print("Dot Product of x and v:")
print(dot_product2)
print("Dot Product of x and y:")
print(dot_product3)

print("\n")

# Concatenate x and y along rows (axis 0)
concat_xy = np.concatenate((x, y), axis=0)
# Concatenate v and w along columns (axis 0, making them 2D first)
v_reshaped = v.reshape(-1, 1)  # Reshape v to a column vector
w_reshaped = w.reshape(-1, 1)  # Reshape w to a column vector
concat_vw = np.concatenate((v_reshaped, w_reshaped), axis=1)
print("Concatenated x and y along rows:")
print(concat_xy)
print("\nConcatenated v and w along columns:")
print(concat_vw)
```

```python
import numpy as np

# Given arrays
A = np.array([[3, 4], [7, 8]])
B = np.array([[5, 3], [2, 1]])

# 1. Prove A * A^-1 = I
A_inv = np.linalg.inv(A)
identity_matrix = np.eye(2)  # Identity matrix of size 2x2

# Multiply A by A^-1
A_A_inv = np.dot(A, A_inv)

print("A * A^-1 (should be identity matrix):")
print(A_A_inv)

# Check if it is close to identity
print("\nCheck if A * A^-1 is identity matrix:", np.allclose(A_A_inv, identity_matrix))

# 2. Prove AB != BA
AB = np.dot(A, B)
BA = np.dot(B, A)

print("\nAB:")
print(AB)

print("\nBA:")
print(BA)

# Check if AB is equal to BA
print("\nCheck if AB == BA:", np.array_equal(AB, BA))

# 3. Prove (AB)^T = B^T * A^T
AB_T = np.transpose(AB)
B_T_A_T = np.dot(np.transpose(B), np.transpose(A))

print("\n(AB)^T:")
print(AB_T)

print("\nB^T * A^T:")
print(B_T_A_T)

# Check if (AB)^T is equal to B^T * A^T
print("\nCheck if (AB)^T == B^T * A^T:", np.array_equal(AB_T, B_T_A_T))
```

```python
import numpy as np

# Coefficient matrix A
A = np.array([[2, -3, 1],
              [1, -1, 2],
              [3, 1, -1]])

# Constant matrix B
B = np.array([-1, -3, 9])

# Calculate the inverse of A
A_inv = np.linalg.inv(A)

# Solve for X using the inverse method
X = np.dot(A_inv, B)

print("The solution is:")
print(f"x = {X[0]}, y = {X[1]}, z = {X[2]}")
```

```python
import numpy as np

# Coefficient matrix A
A = np.array([[2, -3, 1],
              [1, -1, 2],
              [3, 1, -1]])

# Constant matrix B
B = np.array([-1, -3, 9])

# Using np.linalg.inv to find the inverse of A
A_inv = np.linalg.inv(A)

# Solving for X using the formula X = A^-1 * B
X = np.dot(A_inv, B)

# Output the solution
print("The solution is:")
print(f"x = {X[0]}, y = {X[1]}, z = {X[2]}")
```

```python
import time
import numpy as np

# Generate two 1000x1000 matrices with random values using Python lists
matrix1 = [[i + j for j in range(1000)] for i in range(1000)]
matrix2 = [[i - j for j in range(1000)] for i in range(1000)]

# Function to perform matrix multiplication using Python lists
def matrix_multiplication_lists(A, B):
    result = [[0] * len(B[0]) for _ in range(len(A))]
    for i in range(len(A)):
        for j in range(len(B[0])):
            for k in range(len(B)):
                result[i][j] += A[i][k] * B[k][j]
    return result

# Measure time taken for matrix multiplication using Python lists
start_time = time.time()

result_list = matrix_multiplication_lists(matrix1, matrix2)

end_time = time.time()
print(f"Time taken for matrix multiplication using Python lists: {end_time - start_time:.6f} seconds")

# Convert the lists to NumPy arrays for comparison
array1 = np.array(matrix1)
array2 = np.array(matrix2)

# Measure time taken for matrix multiplication using NumPy arrays
start_time = time.time()

result_array = np.dot(array1, array2)

end_time = time.time()
print(f"Time taken for matrix multiplication using NumPy arrays: {end_time - start_time:.6f} seconds")
```
