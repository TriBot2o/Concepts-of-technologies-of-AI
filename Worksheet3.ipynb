```python
total_fare = df['Fare'].sum()
first_class_fare = df[df['Pclass']==1]['Fare'].sum()
second_class_fare = df[df['Pclass']==2]['Fare'].sum()
third_class_fare = df[df['Pclass']==3]['Fare'].sum()

fare_total = [total_fare,first_class_fare,second_class_fare,third_class_fare]
fare_revenue_of_all_classes = [first_class_fare/total_fare,second_class_fare/total_fare,third_class_fare/total_fare]
fare_revenue_of_all_classes
```

```python
import numpy as np
df['age_group'] = np.where(df['Age'] < 18, 'child', np.where((df['Age'] >= 18) & (df['Age'] <= 64), 'adult', 'senior'))
df.age_group.head()
```

```python
total_number_of_passengers = df['PassengerId'].count()
total_number_of_passengers
```

```python
child_passengers = df[df['age_group']=='child']['PassengerId'].count()
adult_passengers = df[df['age_group']=='adult']['PassengerId'].count()
senior_passengers = df[df['age_group']=='senior']['PassengerId'].count()
print(child_passengers,adult_passengers,senior_passengers)
```

```python
proportion_of_passengers_in_each_age_group = [child_passengers/total_number_of_passengers,adult_passengers/total_number_of_passengers,senior_passengers/total_number_of_passengers]
proportion_in_percentage = [proportion*100 for proportion in proportion_of_passengers_in_each_age_group]
proportion_in_percentage
```
